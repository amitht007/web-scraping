```python
names=[]
links=[]
image_links=[]
designations=[]
for alphabet in alphabets:
  not_empty=True
  num=0
  while not_empty:
    num+=1
    url="https://www.nitrkl.ac.in/FacultyStaff/Faculty?pageNumber="+str(num)+"&Filter_Alpha="+alphabet
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,"lxml")
    faculties=soup.find("div",class_="faculty-row").find_all("div",class_="fac-item card")

    if len(faculties)==0:
      print("Broken")
      not_empty=False
    else:
      for faculty in faculties:
        name=names.append(faculty.find("div",class_="card-body").find("a").get_text())
        links.append(faculty.find("div",class_="card-body").find("a").get('href'))
        image_links.append(image_finder(faculty))
        designations.append(faculty.find("div",class_="card-body").find("small").get_text())

print(names)
print("..............")
print(image_links)
print("..............")
print(links)
print("..............")
print(designations)
```

```python
dict={
    "faculty_name":names,

    "faculty_designation":designations,

    "faculty_image":image_links,

    "faculty_link":links
}
data=pd.DataFrame(dict)

data.to_csv("faculty.csv")
```

```python
companys=soup.find_all("div",class_="companyCardWrapper")
ratings=soup.find_all("span",class_="companyCardWrapper__companyRatingValue")

names=[]
ratings=[]
reviews=[]
salaries=[]
interviews=[]
jobs=[]
benefits=[]
photos=[]
high_rated=[]
low_rated=[]
for company in companys:
  names.append(company.find("h2",class_="companyCardWrapper__companyName").text.strip())
  ratings.append(company.find("span",class_="companyCardWrapper__companyRatingValue").text.strip())

  #stats
  stats=company.find_all("span",class_="companyCardWrapper__ActionCount")
  check=0
  for stat in stats:
    if check==0:
      reviews.append(stat.text.strip())
    if check==1:
      salaries.append(stat.text.strip())
    if check==2:
      interviews.append(stat.text.strip())
    if check==3:
      jobs.append(stat.text.strip())
    if check==4:
      benefits.append(stat.text.strip())
    if check==5:
      photos.append(stat.text.strip())
    check+=1

    #rated
    rates=company.find_all("span",class_="companyCardWrapper__ratingValues")
    rate_check=0
  for rate in range(2):
    if rate_check==0:
      high_rated.append(rates[0].text.strip())
    else:
      try:
        low_rated.append(rates[1].text.strip())
      except(IndexError):
        low_rated.append("N/A")
    rate_check+=1
```

```python
company_dict={
    "company_name":names,

    "company_rating":ratings,

    "company_review":reviews,

    "company_salary":salaries,

    "company_interview":interviews,

    "company_job":jobs,

    "company_benefit":benefits,

    "company_photo":photos,

    "company_high_rated":high_rated,

    "company_low_rated":low_rated
}
data=pd.DataFrame(company_dict)
data
```

# THIS IS THE COMPLETE CODE FOR WEB SCRAPING

```python
final=pd.DataFrame()

for i in range(1,3,1):

  url="https://www.ambitionbox.com/list-of-companies?page="+str(i)

  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get(url,headers=headers).text
  soup=BeautifulSoup(webpage,"lxml")

  companys=soup.find_all("div",class_="companyCardWrapper")
  ratings=soup.find_all("span",class_="companyCardWrapper__companyRatingValue")

  names=[]
  ratings=[]
  reviews=[]
  salaries=[]
  interviews=[]
  jobs=[]
  benefits=[]
  photos=[]
  high_rated=[]
  low_rated=[]
  for company in companys:
    names.append(company.find("h2",class_="companyCardWrapper__companyName").text.strip())
    ratings.append(company.find("span",class_="companyCardWrapper__companyRatingValue").text.strip())

    #stats
    stats=company.find_all("span",class_="companyCardWrapper__ActionCount")
    check=0
    for stat in stats:
      if check==0:
        reviews.append(stat.text.strip())
      if check==1:
        salaries.append(stat.text.strip())
      if check==2:
        interviews.append(stat.text.strip())
      if check==3:
        jobs.append(stat.text.strip())
      if check==4:
        benefits.append(stat.text.strip())
      if check==5:
        photos.append(stat.text.strip())
      check+=1

      #rated
      rates=company.find_all("span",class_="companyCardWrapper__ratingValues")
      rate_check=0
    for rate in range(2):
      if rate_check==0:
        try:
          high_rated.append(rates[0].text.strip())
        except(IndexError):
          high_rated.append(np.nan)
      else:
        try:
          low_rated.append(rates[1].text.strip())
        except(IndexError):
          low_rated.append(np.nan)
      rate_check+=1

  company_dict={
      "company_name":names,

      "company_rating":ratings,

      "company_review":reviews,

      "company_salary":salaries,

      "company_interview":interviews,

      "company_job":jobs,

      "company_benefit":benefits,

      "company_photo":photos,

      "company_high_rated":high_rated,

      "company_low_rated":low_rated
  }

  data=pd.DataFrame(company_dict)

  final=final.append(data,ignore_index=True)

```
