```python
from pytube import Playlist,YouTube
```

```python
url="https://www.youtube.com/playlist?list=PLKnIA16_Rmvbr7zKYQuBfsVkjoLcJgxHH"
p = Playlist(url)
```

```python
dict={
    "title":[],
    "length":[],
    "link":[]
}
for x in p:
  yt=YouTube(x)
  length_seconds = yt.length
  minutes, seconds = divmod(length_seconds, 60)
  length_formatted = f"{minutes}:{seconds:02}"
  dict["title"].append(yt.title)
  dict["length"].append(length_formatted)
  dict["link"].append(yt.watch_url)
```

```python
len(dict["title"])
```

```python
data=pd.DataFrame(dict)

data.to_csv("machine_learning.csv")
```
